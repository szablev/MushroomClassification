```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
```

```python
df = pd.read_csv("input.csv")
#df = df.replace(0, np.nan)
#df = df.dropna()

df.head()
```

```python


def recomp(feat_data):
  for column in string_cols:
    cols = pd.get_dummies(feat_data[column], prefix=column)
    feat_data = pd.concat([feat_data,cols], axis=1)

  feat_data = feat_data.drop(columns=string_cols)

  feat_data = feat_data.fillna(-1)

  return feat_data

#feat_data[column], _ = pd.factorize(feat_data[column], sort=True)
```

```python
features = ['cap-diameter','stem-height','stem-width','cap-shape','cap-color','has-ring','ring-type']
label = 'class'

string_cols = ['cap-shape','cap-color','has-ring','ring-type']
int_cols = ['cap-shape','cap-color','has-ring']

feat_data = df[features].copy()
label_data = df[label].copy()

feat_data = recomp(feat_data)
print(feat_data)

X_train, X_test, y_train, y_test = train_test_split(feat_data, label_data, test_size=0.2, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=19, criterion='gini')
clf.fit(X_train, y_train)
prediction = clf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test.values, prediction))
```

```python
##KNeighborsClassifier
from sklearn import neighbors

n_neighbors = 7

clf = neighbors.KNeighborsClassifier(n_neighbors)
clf.fit(X_train, y_train)
clf

prediction = clf.predict(X_test)
print("Accuracy:", metrics.accuracy_score(y_test.values, prediction))
```

```python
##Classification
df_val = pd.read_csv("validation.csv")
# df_val = df_val.replace(0, np.nan)
#df_val = df_val.dropna()
```

```python
feat_data_val = df_val[features].copy()

feat_data_val = recomp(feat_data_val)

prediction = clf.predict(feat_data_val)
prediction
```
